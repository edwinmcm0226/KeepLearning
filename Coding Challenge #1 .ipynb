3. Normalise the stock prices and visualise

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = "{:.4f}".format
plt.style.use("seaborn")
```

```python
# Requirement 1 of this Challenge

symbol = ["TSM","NVDA","0981.HK"]

start = "2015-01-02"
end = "2024-06-30"
```

```python
df = yf.download(symbol, start, end)
df.to_csv("Semicon_OHLC.csv")  # save the raw data
df
```

```python
df.info()
```

```python
close_price = df["Close"]
close_price
```

```python
# Requirement 2 of this Challenge
# To plot TSM's close price

close_price.TSM.dropna().plot(figsize = (15,8), fontsize = 13, logy = False)
plt.legend(fontsize = 13)
plt.show()
```

```python
# Requirement 3 of this Challenge
close_price_normalise = close_price.div(close_price.iloc[0]).mul(100)
close_price_normalise
```

```python
close_price_normalise.dropna().plot(figsize = (15,8), fontsize = 13, logy = True)
plt.legend(fontsize = 13)
plt.show()
```

```python
close_price_normalise.to_csv("close_price_normalise.csv")
```
